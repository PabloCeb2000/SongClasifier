```python
import numpy as np
import pandas as pd
import os
import librosa
from IPython.display import Audio
import matplotlib.pyplot as plt
import librosa.display
from scipy import stats
```

```python
df = pd.read_csv('D:\P\Pr\MClasifier\Rolas\SP_cluster1.csv')
df
```

```python
plt.figure(figsize=(10, 10))
plt.pie(df['cluster'].value_counts(), labels=df['cluster'].value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
class_counts = df.groupby(['cluster', 'Lista']).size().unstack()
class_survival_percentage = class_counts.div(class_counts.sum(axis=1), axis=0) * 100

class_survival_percentage.plot(kind='bar', stacked=True, color=['#5cb85c', '#d9534f'], figsize=(7, 6))
plt.title('Percentage of Passenger Class Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Percentage (%)')
plt.legend(['Did not survive', 'Survived'], title='Survival')
plt.xticks(rotation=0)
plt.show()
```
